The purpose of this file is to answer the following questions about Qutoa ratios:
1. What is the average First_pref:Quota ratio needed to get elected per consistuency
2. What is the minimum needed to get elected?
5. What is the highest First_pref:Quota ratio without being elected? 
3. What is the lowest while still being elected?
4. Is there a trend in First_pref:Quota ratio needed to get elected?

```python
import pandas as pd
import numpy as np

df = pd.read_parquet('ALL_CANDIDATES.parquet')
df.head()
```

```python
_ = df[df.election_type == 'GENERAL']
_.groupby('elected')['first_pref_quota_ratio'].describe()
```

```python
df[(df['first_pref_quota_ratio']>1) &(df['elected'] == False) &(df['election_type'] == 'GENERAL')]
```

```python
cavan = df[(df['constituency'].str.contains('Cavan')) & (df.election_type == 'GENERAL')]
cavan.groupby('elected')['first_pref_quota_ratio'].describe()['50%'].diff().values[-1]
```

```python
general_elections = df[df.election_type == 'GENERAL']
l_o_d = []
for constitency in general_elections.constituency.unique():
    d={}
    print(constitency)
    d['constitency'] = constitency
    election = df[(df['constituency'] == constitency)]
    description = election.groupby('elected')['first_pref_count'].describe()
    print(description)
    for col in description.columns:
        d[col] = description[col][-1]/description[col][0]
    print(
        "The differnece between the median amount of first pref to quota ratio for winners and losers is: ",
        description['50%'][-1]/description['50%'][0]
    )
    print('----------------------------------')
    l_o_d.append(d)

winner_minus_loser = pd.DataFrame(l_o_d)
```
